```python
# import library
import pandas as pd
```

```python
# read in data
netflix = pd.read_csv("netflix_titles.csv")

# Display data
print(netflix.shape)
netflix.head()
```

# Missing Data
### Display the percentage of missing data per column
(netflix.isnull().sum()/netflix.isnull().count()).sort_values(ascending=False)

```python
netflix_v1 = netflix[["type","director"]].dropna()
print(netflix_v1.shape)
(netflix_v1.isnull().sum()/netflix_v1.isnull().count()).sort_values(ascending=False)
```

```python
netflix_v1.groupby("director")["type"].value_counts()
```

```python
netflix_v1.groupby('director')['type'].value_counts().unstack().fillna(0)
```

```python
netflix['date_added'] = pd.to_datetime(netflix['date_added'])
netflix_date = netflix[["country","date_added"]]
print(netflix_date.shape)
print((netflix_date.isnull().sum()/netflix_date.isnull().count()).sort_values(ascending=False))
netflix_date_v1 = netflix_date.dropna()
print((netflix_date_v1.isnull().sum()/netflix_date_v1.isnull().count()).sort_values(ascending=False))
```
